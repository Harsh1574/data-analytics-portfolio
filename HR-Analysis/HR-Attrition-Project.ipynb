```python
df=df.drop_duplicates()
```

```python
df.columns=(
    df.columns
        .str.strip()
        .str.lower()
        .str.replace(r"[^0-9a-zA-Z]+","_",regex=True)
        .str.strip("_")
)
```

```python
group_dept=(df.groupby("department")["attrition"]
            .value_counts(normalize=True)
            .mul(100)
            .round(2)
            .unstack(fill_value=0)
)
group_dept
```

```python
group_ovrtime=(df.groupby("overtime")["attrition"]
   .value_counts(normalize=True)
   .mul(100)
   .round(2)
 .unstack(fill_value=0)
)

group_ovrtime
```

```python
df.to_csv("HR_Attrition_clean.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("Set2")
```

```python
sns.countplot(x="attrition", data=df, palette="Set2")
plt.title("Overall Attrition Distribution")
plt.savefig("overall attrition.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="department",hue="attrition",data=df)
plt.title("Attition by Department")
plt.savefig("attrition by department.png")
plt.show()
```

```python
plt.figure(figsize=(22,10))
sns.countplot(x="jobrole",hue="attrition",data=df)
plt.title("Attition by Job Role")
plt.savefig("attrition by job role.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="gender",hue="attrition",data=df)
plt.title("Attition by Gender")
plt.savefig("attrition by gender.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x="age",hue="attrition",multiple="stack",kde=False, palette="Set2", bins=20)
plt.title("Attition by Age")
plt.savefig("attrition by age.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x="monthlyincome",hue="attrition", palette="Set2")
plt.title("Attition by Monthly Income")
plt.savefig("attrition ny monthly income.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="overtime",hue="attrition", palette="Set2")
plt.title("Attition by Overtime")
plt.savefig("attrition by overtime.png")
plt.show()
```
