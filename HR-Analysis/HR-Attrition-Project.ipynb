```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patheffects as path_effects

# Load cleaned dataset
df = pd.read_csv("HR-Attrition-clean.csv")

# KPI: Overall attrition rate
attrition_rate = (df["attrition"].eq("Yes").mean() * 100).round(1)
overtime_attr = (df[df["overtime"]=="Yes"]["attrition"].eq("Yes").mean()*100).round(1)
entry_attr = (df[df["joblevel"]==1]["attrition"].eq("Yes").mean()*100).round(1)

# Set theme
sns.set_theme(style="whitegrid")
PALETTE = {"Yes": "#FC3E0E", "No": "#1C1CEA"}

fig = plt.figure(figsize=(16, 12), facecolor="#F2EDED")
fig.suptitle("HR Attrition Dashboard", fontsize=40, weight="bold", ha="center")

# --- KPI Card ---

# --- Overall Attrition ---
ax1 = fig.add_axes([0.05, 0.8, 0.25, 0.15])  # x, y, w, h
ax1.axis("off")
ax1.text(0.3, 0.3, "Overall Attrition: ", fontsize=30, ha="center", weight="bold")
ax1.text(0.85, 0.3, f"{attrition_rate}%", fontsize=30, weight="bold", ha="center", color="#FC3E0E")

# --- Overtime Attrition ---
ax1 = fig.add_axes([0.05, 0.8, 0.25, 0.15])  # x, y, w, h
ax1.axis("off")
ax1.text(1.7, 0.3, "Overtime Attrition: ", fontsize=30, ha="center", weight="bold")
ax1.text(2.3, 0.3, f"{overtime_attr}%", fontsize=30, weight="bold", ha="center", color="#FC3E0E")

# --- Entry Attrition ---
ax1 = fig.add_axes([0.05, 0.8, 0.25, 0.15])  # x, y, w, h
ax1.axis("off")
ax1.text(3.25, 0.3, "Entry Attrition: ", fontsize=30, ha="center", weight="bold")
ax1.text(3.75, 0.3, f"{entry_attr}%", fontsize=30, weight="bold", ha="center", color="#FC3E0E")

# --- Attrition by Department ---
ax2 = fig.add_axes([0.05, 0.45, 0.4, 0.3])
sns.countplot(data=df, x="department", hue="attrition", palette=PALETTE, ax=ax2)
ax2.set_title("Attrition by Department", fontsize="20", weight="bold", pad=15)
ax2.set_xlabel("Department", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
ax2.set_ylabel("Employees", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
for container in ax2.containers:
    ax2.bar_label(container, fmt="%d", fontsize=9, padding=3, color="black")

# --- Attrition by Job Role ---
ax3 = fig.add_axes([0.65, 0.45, 0.45, 0.3])
sns.countplot(data=df, y="jobrole", hue="attrition", palette=PALETTE, ax=ax3)
ax3.set_title("Attrition by Job Role", fontsize="20", weight="bold", pad=15)
ax3.set_xlabel("Employees", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
ax3.set_ylabel("Job Role", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
for container in ax3.containers:
    ax3.bar_label(container, fmt="%d", fontsize=9, padding=3, color="black")

# --- Attrition by Overtime ---
ax4 = fig.add_axes([0.05, 0.05, 0.4, 0.25])
sns.countplot(data=df, x="overtime", hue="attrition", palette=PALETTE, ax=ax4)
ax4.set_title("Attrition by Overtime", fontsize="20", weight="bold", pad=15)
ax4.set_xlabel("Overtime", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
ax4.set_ylabel("Employees", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
for container in ax4.containers:
    ax4.bar_label(container, fmt="%d", fontsize=9, padding=3, color="black")

# --- Attrition by Age (stacked with consistent label colors + stroke effects) ---
ax5 = fig.add_axes([0.65, 0.05, 0.45, 0.25])

hue_order = ["No", "Yes"]
PALETTE   = {"Yes": "#FC3E0E", "No": "#1C1CEA"}

sns.histplot(
    data=df, x="age", hue="attrition",
    multiple="stack", bins=20,
    hue_order=hue_order, palette=PALETTE,
    edgecolor="white", alpha=0.95, ax=ax5
)

ax5.set_title("Attrition by Age", fontsize=20, weight="bold")
ax5.set_xlabel("Age", fontsize=15, weight="bold", color="#1C1CEA")
ax5.set_ylabel("Employees", fontsize=15, weight="bold", color="#1C1CEA")

# Label colors & stroke for each category
label_style = {
    "No":   {"color": "white", "path_effects": [path_effects.Stroke(linewidth=1, foreground="black"), path_effects.Normal()]},
    "Yes":  {"color": "black", "path_effects": [path_effects.Stroke(linewidth=1.25, foreground="white"), path_effects.Normal()]}
}

# Loop over bars and add labels
for patch, attr in zip(ax5.patches, 
                       [hue for _ in range(len(ax5.patches)//len(hue_order)) for hue in hue_order]):
    h = patch.get_height()
    if h <= 0:
        continue

    x = patch.get_x() + patch.get_width() / 2.0
    y = patch.get_y() + h / 2.0

    ax5.text(
        x, y, f"{int(h)}",
        ha="center", va="center",
        fontsize=9, weight="bold",
        color=label_style[attr]["color"],
        path_effects=label_style[attr]["path_effects"],
        clip_on=True
    )
# --- Attrition by Gender ---
ax6 = fig.add_axes([0.05, -0.35, 0.4, 0.25])
sns.countplot(data=df, x="gender", hue="attrition", palette=PALETTE, ax=ax6)
ax6.set_title("Attrition by Gender", fontsize="20", weight="bold", pad=15)
ax6.set_xlabel("Gender", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
ax6.set_ylabel("Employees", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
for container in ax6.containers:
    ax6.bar_label(container, fmt="%d", fontsize=9, padding=3, color="black")

# --- Attrition by Monthly Income ---
ax7 = fig.add_axes([0.65, -0.35 , 0.45, 0.25])
sns.boxplot(data=df, x="monthlyincome", hue="attrition", palette=PALETTE, ax=ax7)
ax7.set_title("Attrition by Monthly Income", fontsize="20", weight="bold", pad=15)
ax7.set_xlabel("Monthly Income", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
ax7.set_ylabel("Employees", fontsize=15, labelpad=10, weight="bold", color="#1C1CEA")
for i, artist in enumerate(ax7.artists):
    # get median line (the 5th line in each box)
    median_line = ax7.lines[i*6 + 4]
    median_x = median_line.get_xdata().mean()
    median_y = median_line.get_ydata().mean()
    ax7.text(median_x, median_y, f"{median_y:.0f}", 
             ha="center", va="bottom", fontsize=8, color="black", weight="bold")
    

fig.savefig("Screenshots/summary-dashboard.png", dpi=200, bbox_inches="tight", facecolor=fig.get_facecolor())

plt.show()
```
